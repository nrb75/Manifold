```python
import pandas as pd
import numpy as np

data=pd.read_csv('rules_df.csv')
```

Rules:
    1. start with the highest ranked pair, assign each of these IPs to server '0'
    2. go to next highest pair, check if IP_B exists in any other server, 
        If it does not, then this pair (IP_A and IP_B) can be assigned to the first server that has room. 
        IF IP_B has already been assigned, find the server it has been assigned to and assign IP_A to that server.

```python

serversize=20
serverlist=[]
server={}
ips={}
serverid=0

for i in range(0,len(data)):
    if len(server)==serversize:
        serverlist.append(server)
        server={}
    if data['IP_B'][i] in server and len(server)<=serversize-1 and data['IP_A'][i] not in ips:
        server[data['IP_A'][i]]=serverid
        ips[data['IP_A'][i]]=1
    if data['IP_B'][i] not in server and len(server)<=serversize-2 and data['IP_A'][i] not in ips:
        server[data['IP_A'][i]]=serverid
        server[data['IP_B'][i]]=serverid
        ips[data['IP_A'][i]]=1
        ips[data['IP_B'][i]]=1
    if data['IP_B'][i] not in server and len(server)==(serversize-1) and data['IP_A'][i] not in ips: #if there is not enough room for the pair, we need to start a new server even if it is not full
        serverlist.append(server)
        server={}
        serverid=serverid+1
        server[data['IP_A'][i]]=serverid
        server[data['IP_B'][i]]=serverid
        ips[data['IP_A'][i]]=1
        ips[data['IP_B'][i]]=1     
    
#when I add in restriction where IP_B also not in ips, it will not find them
```
